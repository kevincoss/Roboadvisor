```python
import yfinance as yf
import pandas as pd

# S&P 500 지수의 티커
ticker = '^GSPC'

# 날짜 범위 설정
start_date = '2003-01-01'
end_date = '2023-01-01'

# 데이터 불러오기
data = yf.download(ticker, start=start_date, end=end_date)

# 일간 수익률 계산
daily_returns = data['Adj Close'].pct_change()

# 연간화 변동성 계산 (일일 변동성 * sqrt(252))
annualized_volatility = daily_returns.std() * (252 ** 0.5)

# 결과 출력
print("S&P 500의 20년 연간화 변동성:")
print(annualized_volatility)
```

```python
import yfinance as yf
import pandas as pd
import numpy as np

# 각 자산에 해당하는 ETF 티커
assets = {
    'Stock': 'SPY',    # S&P 500 ETF
    'Long_Term_Bond': 'TLT',  # 미국 장기 국채 ETF
    'Mid_Term_Bond': 'IEF',  # 미국 중기 국채 ETF
    'Gold': 'GLD',    # 금 ETF
    'Commodities': 'DBC'  # 원자재 ETF
}

# 올웨더 포트폴리오의 비율
weights = {
    'Stock': 0.30,
    'Long_Term_Bond': 0.40,
    'Mid_Term_Bond': 0.15,
    'Gold': 0.075,
    'Commodities': 0.075
}

# 20년간의 데이터 가져오기
data = yf.download(list(assets.values()), start="2003-01-01", end="2023-01-01")['Adj Close']

# 수익률 계산
returns = data.pct_change()

# 포트폴리오 수익률 계산
portfolio_returns = returns.dot(list(weights.values()))

# 누적 수익률 계산
cumulative_returns = (1 + portfolio_returns).cumprod() - 1

# 결과 출력
cumulative_returns.plot(title='올웨더 포트폴리오 20년간 누적 수익률')
plt.xlabel('Year')
plt.ylabel('Cumulative Return')
plt.show()
```
