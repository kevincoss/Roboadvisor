```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
tickers = ['AAPL', 'F', 'AMZN', 'GE', 'TSLA']

# 빈 데이터프레임 생성
pxclose = pd.DataFrame()

# 각 티커에 대해 반복하여 데이터를 가져오고 'Adj Close' 컬럼 선택
for t in tickers:
    ticker_data = yf.Ticker(t)
    hist_data = ticker_data.history(start='2019-01-01', end='2019-12-31')
    pxclose[t] = hist_data['Close']  # 'Adj Close' 대신 'Close' 사용

pxclose
```

```python
ret_daily = pxclose.pct_change()
ret_daily
```

```python
ret_annual = ret_daily.mean() * 250
ret_annual
```
