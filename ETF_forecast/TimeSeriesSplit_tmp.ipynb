```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold

# 데이터 생성
X = np.arange(100)
y = np.arange(100, 200)

# KFold 객체 생성
n_splits = 5
kf = KFold(n_splits=n_splits)

# 시각화 설정
fig, ax = plt.subplots()

# 색상 설정
colors = ['#ff7f0e', '#1f77b4']  # 주황, 파랑
test_color = colors[0]
train_color = colors[1]

# KFold 시각화
for i, (train_index, test_index) in enumerate(kf.split(X)):
    # 훈련 데이터 표시
    ax.scatter(train_index, [i] * len(train_index), c=train_color, marker='s', edgecolor='none', label='Training set' if i == 0 else "")
    # 테스트 데이터 표시
    ax.scatter(test_index, [i] * len(test_index), c=test_color, marker='s', edgecolor='none', label='Testing set' if i == 0 else "")

# 레이블과 범례 추가
ax.legend(loc='upper right')
ax.set_title('KFold')
ax.set_xlabel('Sample index')
ax.set_ylabel('CV iteration')

# 배경 스타일을 두 번째 이미지와 유사하게 설정
fig.patch.set_facecolor('#EAEAF2')
ax.set_facecolor('#EAEAF2')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold

# 데이터 생성
X = np.arange(100)
y = np.arange(100, 200)

# KFold 객체 생성
n_splits = 5
kf = KFold(n_splits=n_splits)

# 테스트 데이터의 앞과 뒤에서 제외할 데이터 수 설정
n = 2

# 시각화 설정
fig, ax = plt.subplots()

# 색상 설정
colors = ['#ff7f0e', '#1f77b4']  # 주황, 파랑
test_color = colors[0]
train_color = colors[1]

# KFold 시각화 및 데이터 분할
for i, (train_index, test_index) in enumerate(kf.split(X)):
    # 테스트 데이터 앞과 뒤에서 n개 제외
    if len(test_index) > 2 * n:
        test_index = test_index[n:-n]
    elif len(test_index) > n:
        test_index = test_index[n:] if train_index[-1] < test_index[0] else test_index[:-n]

    # 훈련 데이터 표시
    ax.scatter(train_index, [i] * len(train_index), c=train_color, marker='s', edgecolor='none', label='Training set' if i == 0 else "")
    # 테스트 데이터 표시
    ax.scatter(test_index, [i] * len(test_index), c=test_color, marker='s', edgecolor='none', label='Testing set' if i == 0 else "")

    # 출력
    print(f"Fold: {i}")
    print("TRAIN indices:", train_index)
    print("TEST indices:", test_index)
    print("\n")

# 레이블과 범례 추가
ax.legend(loc='upper right')
ax.set_title('KFold')
ax.set_xlabel('Sample index')
ax.set_ylabel('CV iteration')

# 배경 스타일을 두 번째 이미지와 유사하게 설정
fig.patch.set_facecolor('#EAEAF2')
ax.set_facecolor('#EAEAF2')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold

# 데이터 생성
X = np.arange(100)
y = np.arange(100, 200)

# KFold 객체 생성
n_splits = 5
kf = KFold(n_splits=n_splits)

# 테스트 데이터의 앞과 뒤에서 제외할 데이터 수 설정
n_start = 6  # 테스트 데이터 앞에서 제외할 개수
n_end = 2    # 테스트 데이터 뒤에서 제외할 개수

# 시각화 설정
fig, ax = plt.subplots()

# 색상 설정
colors = ['#ff7f0e', '#1f77b4']  # 주황, 파랑
test_color = colors[0]
train_color = colors[1]

# KFold 시각화 및 데이터 분할
for i, (train_index, test_index) in enumerate(kf.split(X)):
    # 테스트 데이터 앞과 뒤에서 n_start, n_end개 제외
    if len(test_index) > (n_start + n_end):
        test_index = test_index[n_start:-n_end]
    elif len(test_index) > n_start:
        test_index = test_index[n_start:] if train_index[-1] < test_index[0] else test_index[:-n_end]
    elif len(test_index) > n_end:
        test_index = test_index[:-n_end]

    # 훈련 데이터 표시
    ax.scatter(train_index, [i] * len(train_index), c=train_color, marker='s', edgecolor='none', label='Training set' if i == 0 else "")
    # 테스트 데이터 표시
    ax.scatter(test_index, [i] * len(test_index), c=test_color, marker='s', edgecolor='none', label='Testing set' if i == 0 else "")

    # 출력
    print(f"Fold: {i}")
    print("TRAIN indices:", train_index)
    print("TEST indices:", test_index)
    print("\n")

# 레이블과 범례 추가
ax.legend(loc='upper right')
ax.set_title('KFold')
ax.set_xlabel('Sample index')
ax.set_ylabel('CV iteration')

# 배경 스타일을 두 번째 이미지와 유사하게 설정
fig.patch.set_facecolor('#EAEAF2')
ax.set_facecolor('#EAEAF2')

plt.show()
```

```python
import numpy as np
from sklearn.model_selection import KFold

def custom_split(X, y, n_splits, n_start, n_end):
    kf = KFold(n_splits=n_splits)
    splits = []

    for train_index, test_index in kf.split(X):
        # 테스트 데이터 앞과 뒤에서 n_start, n_end개 제외
        if len(test_index) > (n_start + n_end):
            test_index = test_index[n_start:-n_end]
        elif len(test_index) > n_start:
            test_index = test_index[n_start:] if train_index[-1] < test_index[0] else test_index[:-n_end]
        elif len(test_index) > n_end:
            test_index = test_index[:-n_end]

        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        splits.append((X_train, X_test, y_train, y_test))
    
    return splits

# 데이터 생성
X = np.arange(100)
y = np.arange(100, 200)

# 설정 값
n_splits = 5
n_start = 3  # 테스트 데이터 앞에서 제외할 개수
n_end = 2    # 테스트 데이터 뒤에서 제외할 개수

# 데이터 분할 함수 호출
split_data = custom_split(X, y, n_splits, n_start, n_end)

# 결과 출력
for i, (X_train, X_test, y_train, y_test) in enumerate(split_data):
    print(f"Fold: {i}")
    print("X_train:", X_train)
    print("X_test:", X_test)
    print("y_train:", y_train)
    print("y_test:", y_test)
    print("\n")
```
