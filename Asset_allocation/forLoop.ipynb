```python
# 각 자산의 비율0%~100%까지 5%씩 지정. 총 20개의 가능성 가질 수 있음(100/5=20)
n_ports = 20

for i in range(1, n_ports+1):
    for j in range(i+1, n_ports):
        for k in range(i+2, n_ports-1):
            print(i, j, k)
```

```python
# 자산이 7개, 비중 기준은 5% 일때, 생성될 포트폴리오의 경우의 수
# 각 자산의 비중이 0%에서 100%까지 5%씩 증가할 때, 7개 자산으로 구성될 수 있는 포트폴리오의 경우의 수를 계산합니다.

# 자산의 수
n_assets = 7

# 자산 비중 단위(5%)에 따른 비중의 가능한 수 (0%에서 100%까지 5% 단위이므로 총 21가지 경우)
n_steps = 21

# 포트폴리오의 경우의 수 계산
# 각 자산에 대해 독립적으로 비중을 선택할 수 있으므로, 경우의 수는 n_steps의 n_assets 제곱입니다.
# 단, 총합이 100%를 초과하는 경우는 제외해야 하지만, 이 계산에서는 모든 가능한 조합을 단순 계산합니다.
portfolio_cases = n_steps ** n_assets

portfolio_cases
```

```python
# A와 B 두 자산이 있을 때, A의 비중이 0%에서 100%까지 5% 단위로 증가할 때 B의 비중을 계산하여 유효한 포트폴리오의 경우의 수를 계산합니다.
import numpy as np

# A 자산의 가능한 비중
a_weights = np.arange(0, 101, 5)

# B 자산의 비중은 A의 비중에 따라 자동으로 결정됩니다(총합이 100%이므로).
b_weights = 100 - a_weights

# 유효한 포트폴리오의 경우의 수
portfolio_cases = len(a_weights)

portfolio_cases, a_weights, b_weights
```

```python
from scipy.special import comb

# n = 21 (0%에서 100%까지 5% 단위로 나눈 경우의 수, 0% 포함)
# r = 7 (자산의 개수)
n = 20
r = 7

# 조합 공식을 사용하여 경우의 수 계산
portfolio_cases = comb(n+r-1, r-1)

portfolio_cases
```
