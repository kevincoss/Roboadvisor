```python
# DART에 공시된 회사 리스트 불러오기
corp_list = dart.get_corp_list()
```

```python
# KB금융 검색
kb_finance = corp_list.find_by_corp_name('KB금융', exactly=True)[0]

# 2010년부터 연간 연결재무제표 불러오기
fs = kb_finance.extract_fs(bgn_de='20100101', report_tp='quarter') # Q: 분기별

# 재무제표 검색 결과를 엑셀파일로 저장 (기본저장위치: 실행폴더/fsdata)
fs.save()
```
